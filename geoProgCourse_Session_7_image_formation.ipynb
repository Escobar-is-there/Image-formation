# **Structure of a RGB image**

---

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(80)

# Customize dimensions
width, height, depth = 5, 4, 3  # Adjust width, height, and number of color channels

# Create a random RGB array
rgb_array = np.random.randint(0, 256, size=(height, width, depth))

# Display the image with proper extent
plt.imshow(rgb_array, extent=[0, width, 0, height])

# Define grid with axis='both' to display grid lines on both x and y axes
plt.grid(True, which='both', color='black', linewidth=1)

# Show the tick count on both x and y axes
plt.xticks(np.arange(0, width, 1))
plt.yticks(np.arange(0, height, 1))

# Save the figure
plt.savefig("rgbfff_grid.png")
# print(rgb_array)
# Slice the red channel
red_channel = rgb_array[:2, :4, 0]
green_channel = rgb_array[:, :, 1]
blue_channel = rgb_array[:, :, 2]

# Print the red channel
print(red_channel,"Red band \n")
print(green_channel,"Green band \n")
print(blue_channel,"blue band")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Read the image
img = mpimg.imread('/content/drive/MyDrive/RGB Band separation/wallpaperflare.com_wallpaper (3).jpg')  # Replace 'image.jpg' with the path to your image file

# Separate the RGB channels
red_channel = img[:,:,0]
green_channel = img[:,:,1]
blue_channel = img[:,:,2]

# Display the original image and the separated channels
plt.figure(figsize=(10, 6))

# Display the original image
plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title('Original RGB Image')

# Display the red channel
plt.subplot(2, 2, 2)
plt.imshow(red_channel, cmap='Reds')
plt.title('Red Channel')

# Display the green channel
plt.subplot(2, 2, 3)
plt.imshow(green_channel, cmap='Greens')
plt.title('Green Channel')

# Display the blue channel
plt.subplot(2, 2, 4)
plt.imshow(blue_channel, cmap='Blues')
plt.title('Blue Channel')

# Adjust spacing between subplots
plt.subplots_adjust(hspace=0.3, wspace=0.2)
plt.savefig("separated_R_G_B_Bangladesh.png")

plt.show()
```
